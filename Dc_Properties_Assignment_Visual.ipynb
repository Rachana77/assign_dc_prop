```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
import seaborn as sns

filename = "DC_Properties.csv"
df = pd.read_csv(filename)

df['PRICE'].fillna(df['PRICE'].mean(), inplace=True)

df.dropna(subset=['ZIPCODE','WARD'], inplace=True)
```

```python
df.dropna(subset=['AYB'], inplace=True)
df.dropna(subset=['NUM_UNITS'], inplace=True)
```

```python
sns.lmplot(x='AYB', y='NUM_UNITS', data=df)
```
